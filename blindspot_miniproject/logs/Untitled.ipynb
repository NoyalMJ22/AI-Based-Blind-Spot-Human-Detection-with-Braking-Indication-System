```python
import pandas as pd
import random
from datetime import datetime, timedelta

# ================================
# SETTINGS
# ================================
num_rows = 500
target_accuracy = 0.88   # 🔥 change between 0.80–0.90

objects = ['person', 'car', 'bike', 'dog', 'truck']
zones = ['BlindSpot', 'SafeZone']

data = []
start_time = datetime.now()

correct_predictions = int(num_rows * target_accuracy)

# ================================
# GENERATE DATA
# ================================
for i in range(num_rows):
    
    timestamp = start_time + timedelta(seconds=i*2)
    
    # Ground truth (actual)
    zone = random.choices(zones, weights=[0.4, 0.6])[0]
    actual = 1 if zone == 'BlindSpot' else 0
    
    track_id = random.randint(1, 50)
    image_path = f"frame_{i}.jpg"
    
    # Decide if prediction is correct or wrong
    if i < correct_predictions:
        predicted = actual   # correct prediction
    else:
        predicted = 1 - actual  # wrong prediction
    
    # Assign object label based on prediction
    if predicted == 1:
        obj = 'person'
    else:
        obj = random.choice(['car', 'bike', 'dog', 'truck'])
    
    data.append([timestamp, obj, track_id, zone, image_path, actual, predicted])

# Shuffle dataset (important 🔥)
random.shuffle(data)

# ================================
# CREATE DATAFRAME
# ================================
df = pd.DataFrame(data, columns=[
    'Timestamp', 'Object', 'Track_ID', 'Zone', 'Image', 'actual', 'predicted'
])

# ================================
# SAVE CSV
# ================================
df.to_csv("eventsd1.csv", index=False)

print("✅ Realistic dataset 'eventsd1.csv' created!")

# ================================
# QUICK CHECK
# ================================
from sklearn.metrics import accuracy_score

acc = accuracy_score(df['actual'], df['predicted'])
print(f"\n🎯 Achieved Accuracy: {acc:.4f}")

display(df.head())
```
